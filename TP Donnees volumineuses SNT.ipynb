!!! question Question 10
Il est possible aussi de trier les résultats. Trions la dernière recherche : les résultats sont triés par le nombre d'enfants portant ces prénoms. Exécutez la cellule ci-dessous
!!!

````python
recherche = prenoms.loc[(prenoms['dpt'] == "14") & (prenoms['annais'] == "2019"), :]
tri = recherche.sort_values(by = "nombre")
print(tri)
````
