## 1. Qu'est-ce que Capytale ?
    

</div>

<div style="border-left:3px solid #007f99; border-radius:2pt; color:#007f99; padding:16px; background-color: #e8f0f5">

<img src="https://raw.githubusercontent.com/reiamnat/seminaireSNT/main/outils.png" alt="Alternative text"  width = "400"/>
    
</div>

<div style="border-left:3px solid #007f99; border-radius:2pt; color:#007f99; padding:16px; background-color: #e8f0f5">

**Capytale :**
* est un service proposé par l'académie de Paris en lien avec la Drane Orléans-Tours,
* permet la création et le **partage entre enseignants et élèves** d'activités utilisant entre autres les langages Python, HTML/CSS et micropython

Le système fonctionne en ligne sans installation et offre une interface commune pour différentes activités d’apprentissage, sur tout périphérique disposant d’un navigateur : ordinateur, tablette ou téléphone.
</div>

<div style="border-left:5px solid #ff9033; border-radius:2pt; color:#ff9033; padding:16px; background-color: #007f99">


## 2. Comment se connecter à Capytale ?

</div>

<div style="border-left:3px solid #007f99; border-radius:2pt; color:#007f99; padding:16px; background-color: #e8f0f5">

L'accès à Capytale se fait, pour les enseignants comme pour les élèves, via l'ENT, sans aucune création de compte.

    
Si l'icône Capytale <img src="https://raw.githubusercontent.com/reiamnat/seminaireSNT/main/iconeCapytale.png" alt="icone" style="display : inline; width : 100px"/> n'est pas visible, demander au référent ENT de l'établissement d'attribuer le connecteur aux enseignants et aux élèves.
    
 </div>

<div style="border-left:5px solid #ff9033; border-radius:2pt; color:#ff9033; padding:16px; background-color: #007f99">


## 3. Quels types d’activités peut-on créer ?

</div>

<div style="border-left:3px solid #007f99; border-radius:2pt; color:#007f99; padding:16px; background-color: #e8f0f5">
    
  <img src="https://raw.githubusercontent.com/reiamnat/seminaireSNT/main/activitesSNT.png" alt="Alternative text"  width = "600"/>

Capytale propose la création :

* d'activités de programmation en python, en mode script/console ou dans un notebook, en intégrant les outils du site Basthon.fr ;
* d'activités de création de contenus web en HTML, CSS (et JavaScript). Chaque zone peut être masquée ou verrouillée en lecture seule selon le besoin pédagogique ;
* d'activités portant sur la programmation de la carte micro:bit (disponible à la rentrée)

Chaque activité peut :

* être partagée avec les élèves,
* être associée à des mots-clés personnels,
* être publiée dans la bibliothèque
* être co-gérée par d’autres enseignants 
    
    
   
 </div>   
 
 

<div style="border-left:5px solid #ff9033; border-radius:2pt; color:#ff9033; padding:16px; background-color: #007f99">


## 4. Comment partager une activité avec les élèves ? Quelles fonctionnalités sont disponibles ?

</div>

<div style="border-left:3px solid #007f99; border-radius:2pt; color:#007f99; padding:16px; background-color: #e8f0f5">
 
Si l'on souhaite partager une activité avec des élèves, un paramètre (activé par défaut lors de la création des activités) génère un code du type `4ad7-1749801` qu'il suffit de communiquer aux élèves.

Lorsque l’élève se connecte (via l’ENT) à Capytale, il saisit le code de l’activité : une copie personnelle de l’activité est affichée.    
    Il peut modifier le contenu pour répondre à la consigne et enregistrer (sur le cloud de Capytale), autant de fois qu’il le souhaite.    
    Lorsque l’élève considère que son travail est terminé, il clique sur le bouton « Rendre ce travail » : il ne peut alors plus le modifier.
</div>
    

<div style="border-left:3px solid #007f99; border-radius:2pt; color:#007f99; padding:16px; background-color: #e8f0f5">    
L’enseignant voit de son côté le statut de chaque copie d’élève et peut accéder à son contenu à tout moment :<br>


* **En cours** : l’élève peut modifier son travail, qu’il n’a pas encore rendu
* **Rendue** : la copie n'est plus modifiable par l’élève mais elle n’est pas encore corrigée
* **Corrigée** : la copie n'est plus modifiable par l’élève et l’enseignant indique qu’il l’a corrigée.

Chaque copie contient deux cadres permettant à l’enseignant de renseigner une appréciation et une évaluation. L’enseignant peut également redonner la main à l’élève si la copie a été rendue de façon prématurée.
 
    
</div>
