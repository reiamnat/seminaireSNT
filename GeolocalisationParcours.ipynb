3. Pour créer une carte permettant de visualiser le lieu de prise de la photo, nous allons utiliser le module folium.
    
Pour utiliser celui-ci, il faut impérativement exécuter la cellule ci-dessous.

```python
import folium                                # import du module pour générer une carte
from pyroutelib3 import Router 
```

```python
              # import des fonctionnalités pour générer un itinéraire

def creation_carte(latitude_centre, longitude_centre, zoom):
    centre = latitude_centre, longitude_centre
    # Génération de la carte
    return folium.Map(centre, zoom_start=zoom)
```

```python
latitude_centre, longitude_centre = (48.079584845875885, 7.357623229009501)
zoom = 15

carte = creation_carte(latitude_centre, longitude_centre, zoom)
carte
```

```python
def marqueur_image(carte, latitude, longitude, url_image, message=""):
    icon = folium.features.CustomIcon(icon_image=url_image, icon_size=(30, 30))
    folium.Marker([latitude, longitude], tooltip=message, icon=icon).add_to(carte) 
    
```

```python


def itineraire(Carte, Depart, Arrivee, mode, couleur):
    """
    ajoute sur carte  un itinéraire avec :
    Depart : coordonnées GPS du point de départ
    Arrivee : coordonnées GPS du point d'arrivée
    mode : mode de transport ("car","cycle","foot","horse","tram","train")
    """
    
     
    try:
        # Génération de l'itinéraire avec le mode de transport choisi
        router = Router(mode)
        D = router.findNode(*Depart)
        A = router.findNode(*Arrivee)
        routeLatLons=[Depart, Arrivee]
        status, route = router.doRoute(D, A)
        if status == 'success': routeLatLons = list(map(router.nodeLatLon, route))
    
        # Tracé de l'itinéraire sur la carte
        folium.PolyLine(routeLatLons, color=couleur, weight=2.5, opacity=1).add_to(Carte)
    
    except:
        # Si l'itinéraire ne peut pas être généré: affichage d'un message d'erreur
        print("impossible de générer l'itinéraire demandé")


```

```python
itineraire(carte, (48.079584845875885, 7.357623229009501) , (48.07103371359589, 7.353385338720863) , "car")
carte
```

```python
itineraire(carte, (48.07103371359589, 7.353385338720863) , (48.08822231782525, 7.37523897770312) , "car")
carte
```
