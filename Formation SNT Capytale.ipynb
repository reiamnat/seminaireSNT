### 3.2 Cycle de vie d’une copie (partage/bloquée/réouverte)
    
Trois états possibles pour une copie : 
    
* **en cours** :  obligatoire tant que l'élève a besoin de travailler sur sa copie.   
    La copie de l'élève possède le bouton : <img style="display : inline" src="https://raw.githubusercontent.com/reiamnat/seminaireSNT/main/tnr.png">
    
* **rendue** : L'élève ne peut plus modifier sa copie.  
    La copie de l'élève possède le bouton : <img style="display : inline" src="https://raw.githubusercontent.com/reiamnat/seminaireSNT/main/tr.png">

* **corrigée** : L'élève ne peut plus modifier sa copie.  
    La copie de l'élève possède le bouton  : <img style="display : inline" src="https://raw.githubusercontent.com/reiamnat/seminaireSNT/main/tc.png">
    
Dans tous les cas, l'enseignant peut visualiser le contenu de chaque copie. Il peut modifier lui-même un état, notamment pour indiquer qu'une copie est corrigée, ou redonner la main à un élève qui a rendu sa copie ; ceci est aussi possible dans la fenêtre des vues  :<br>

### 3.3 Possibilité de cloner une activité/modifier les paramètres

Une fois votre activité créée, plusieurs options vous sont proposées :
    
    
    
![modifier.png](https://raw.githubusercontent.com/reiamnat/seminaireSNT/main/modifier.png)

    
* Pour modifier votre activité, choisir **paramètres** : vous retrouverez la page de création de celle-ci.
* Pour faire une copie indépendante de votre activité, choisir **cloner** (cela génère un nouveau code).
* Pour stopper la distribution de l'activité, choisir **bloquer la distribution**. Les élèves n'auront plus accès à cette activité. 

Sachez qu'un élève est propriétaire de sa copie, il est le seul à pouvoir la supprimer.

### 3.5 Ré-initialiser une copie d’élève
 
Comment ré-initialiser l'énoncé d'origine si l'élève a :
- ouvert une copie avec un code
- effacé une partie du script (script-console) ou une cellule (notebook) de l'énoncé ?
    
Le bouton : <button class='fa fa-repeat'></button> permet de revenir au dernier énoncé enregistré par l'enseignant.  
Un message d'alerte apparaîtra pour les élèves :   
« Que voulez-vous faire ? »  Restaurer une sauvegarde ou Revenir à la version de l'enseignant
</div>

<div style="border-left:3px solid #007f99; border-radius:2pt; color:#007f99; padding:16px; background-color: #e8f0f5">

### 3.6 Modifier un énoncé partagé

Comment modifier un énoncé que des élèves ont déjà consulté ? :
* Faire la correction côté enseignant et enregistrer.
* Demander aux élèves possédant une copie de l'ancienne version de ré-initialiser le sujet initial (voir 3.5), avec perte du travail éventuellement déjà réalisé, d'enregistrer, de quitter le travail en cours, puis de recharger le sujet : les modifications de l'enseignant sont alors prises en compte.    

    
Dans le cas d'une activité html, les modifications sont éffectuées automatiquement dans les copies des élèves.
</div>

### Exemple

Lors de la création d'une activité, cocher « Joindre des fichiers annexes ... ».
    
les formats de fichiers autorisés pour l'hébergement dans le serveur de Capytale sont : `py, npy, pyc, txt, sql, db, sqlite, csv, png, jpg, bmp, svg, gif, html, pdf, css, zip, tar, tgz, gz, aiff, wav, ml, ttf, json`. 
    

    
- Pour les images : utiliser les méthodes d'insertion Markdown ou html en copiant l'url de l'image

On l'utilise comme une bibliothèque python avec son nom d'origine : 
```python
from ma_bibliotheque import *  
```

    
- Pour cacher un fichier python aux élèves, il suffit avant de le joindre de le renommer en ajoutant un tiret bas avant son nom.

 Par exemple, pour cacher le contenu de  `ma_bibliotheque.py`, il suffit de le renommer `_ma_bibliotheque.py` avant de le joindre puis de l'utiliser comme ceci :
```python
from _ma_bibliotheque import *  
```

### Exemple
    
Le fichier `_visualisation.py` contient la définition de la fonction `dessiner` que l'on veut faire utiliser aux élèves
    
```python
from ipythonblocks import BlockGrid


def dessiner(mat):
    nb_lignes = len(mat)
    nb_colonnes = len(mat[0])
    grille = BlockGrid(nb_colonnes, nb_lignes, block_size=30)
    for ligne in range(nb_lignes):
        for colonne in range(nb_colonnes):
            grille[ligne, colonne] = mat[ligne][colonne]
    return grille    
    
```

### 5.2 Extension RISE
Il est possible de tranformer le notebook en **diaporama interactif** (les cellules de code restent exécutables) avec l'extension **RISE**. 

- Dans le menu, cliquer sur `Affichage` > `Barre d'outils de cellule` > `Diaporama`.  

- Cliquer sur la liste déroulante "Type de diapo" de la première cellule, et sélectionner `Diapo`. 

- Pour la cellule suivante les options sont : 
    - `Diapo`, pour que cette cellule apparaisse **sur une nouvelle diapo** (à droite de la précédente).
    - `Sous-diapo` pour que cette cellule apparaisse seule, **sur une nouvelle diapo**, en-dessous de la première diapo.
    - `Extrait` pour que cette cellule apparaisse sur la première diapo, en dessous de la première cellule, **après l'appui sur la touche <kbd>Espace</kbd>** (affichage progressif de cellules)
    - `-`pour que cette cellule apparaisse sur la première diapo, en dessous de la première cellule (et en même temps qu'elle) (affichage simultané)
    - `Sauter` pour ne pas faire apparaitre cette cellule. 
    
**Chaque cellule** est ainsi à configurer.
    
Une fois le travail terminé, `Affichage` > `Barre d'outils de cellule` > `Aucun`.

En cliquant sur l'icône <button class="btn btn-default" title="Enter/Exit RISE Slideshow" id="RISE" data-jupyter-action="RISE:slideshow"><i class="fa-bar-chart fa"></i></button> présente dans la barre d'outils, la première diapo apparait.
Le <kbd>?</kbd> fait apparaitre un menu avec toutes les commandes.
