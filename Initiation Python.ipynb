!!! info Info
Voici quelques méthodes disponibles pour contrôler la tortue :

* `forward(nombre_de_pas)` : la tortue avance du nombre de pas indiqué en paramètre sachant qu'un pas vaut 1 pixel.
* `backward(nombre_de_pas)` : la tortue recule du nombre de pas indiqué en paramètre.
* `left(angle)` : la tortue tourne vers la gauche du nombre de degrés indiqué en paramètre. 
* `right(angle)` : la tortue tourne vers la droite du nombre de degrés indiqué en paramètre.
* `penup()` : la tortue lève le crayon autrement dit elle ne laisse aucune trace lorsqu'elle se déplace
* `pendown()` : la tortue pose le crayon autrement dit elle laisse une trace lorsqu'elle se déplace.
* `goto(x, y)` : la tortue se déplace vers le point de coordonnées `(x, y)` sans changer d'orientation.

**Par défaut, la tortue est orienté horizontalement gauche-droite.**
!!!

!!! example Exemples
* **avancer**
```python 
forward(50)
```
* **reculer**
```python 
backward(50)
```
* **tourner à gauche**
```python 
left(90)
```
* **tourner à droite**
```python 
right(90)
```

Avec Scratch, vous avez sans doute dessiné un triangle équilatéral en utilisant les instructions suivantes :



!!! info Info
En python, pour répéter plusieurs fois une ou des instructions, on utilise une boucle. Voyons sur un exemple : 

**Le code**
```python
for compteur in range(3) :
    forward(50)
    left(120)
```
**L'explication**
Cette boucle permet de répéter 3 fois : "avancer de 50 pas et tourner de 120° à gauche" 

L'indentation (décalage à droite de 4 espaces) permet de délimiter le bloc de contenu de la boucle : ce qui est indenté sera répété 3 fois.

`compteur` est une variable qui va contenir successivement les valeurs 0, 1 et 2 (puisqu'on à 3). Vous pouvez donner un autre nom à cette variable.      
En mathématiques, l'usage veut qu'on l'appelle souvent `i` ou `j` ou `k`.
!!!

!!! warning Attention
Ne pas oublier le double point en fin de ligne pour débuter un bloc indenté.
!!!

 
### Les variables

!!! info Info
Une variable sert à stocker des informations. Le contenu d'une variable peut être modifié.
!!!

!!! example Exemples

* **Affecter une valeur à une variable**
```python
longueur = 50
```
Le nombre 50 est stocké dans la variable dont le nom est `longueur`. Pour utiliser la valeur dans une instruction, on donne le nom de la variable.

* **Ajouter une valeur à une variable**
```python
longueur = longueur + 10
```
Cette instruction est exécutée de la façon suivante : la machine récupère la valeur contenue dans la variable `longueur`, lui ajoute 10 et le résultat est stocké à nouveau dans la variable `longueur`.  
!!!
